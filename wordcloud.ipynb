```python
import os
from os import path
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```python
# Read the whole text.
text = open(path.join(d, 'p2p.txt')).read()

stopwords = set(STOPWORDS)
stopwords.add("system")
stopwords.add("systems")
stopwords.add("SAP")
stopwords.add("processes")
stopwords.add("within")
stopwords.add("P2P")



wc = WordCloud(background_color="white", max_words=50,
               stopwords=stopwords, contour_width=3, contour_color='steelblue',max_font_size=60)

# Generate a word cloud image
wc.generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
# wc(max_font_size=65).generate(text)
# plt.figure()
# plt.imshow(wc, interpolation="bilinear")
# plt.axis("off")
# plt.show()
```
